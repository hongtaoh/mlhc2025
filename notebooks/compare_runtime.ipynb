Compare runtime. 

```python
import pandas as pd

filename = '../../all_results.csv'

# Load the data
df = pd.read_csv(filename)

# Step 1: Overall avg runtime per algo (used for sorting)
algo_mean = df.groupby('algo')['runtime'].mean().sort_values()

# Step 2: Avg runtime per algo and J
grouped = df.groupby(['algo', 'J'])['runtime'].mean().reset_index()

# Step 3: Pivot the table: rows = algo, cols = J
pivot_table = grouped.pivot(index='algo', columns='J', values='runtime')

# Step 4: Sort rows by overall mean runtime
pivot_table = pivot_table.loc[algo_mean.index]/60

# Step 5: Round to 2 decimal places and export as LaTeX
pivot_table = pivot_table.round(2)

# Save to LaTeX file
pivot_table.to_latex('algo_runtime_table.tex', float_format="%.2f")

pivot_table
```
